```python
from pyspark.sql import SparkSession, functions as F
```

```python
spark = (
    SparkSession.builder
    .appName("tmdb_raw_to_structured")
    .master("local[*]")   # remove or change when running on cluster
    .getOrCreate()
)
```

```python
df = spark.read.parquet('/home/dphung/projects/tmdb/data/silver/tmdb/movies/part-00000-73dc9362-e7e4-401e-94dd-6c07fc612f99-c000.snappy.parquet')
```

```python
df.show()
```

```python
input_path = "/home/dphung/projects/tmdb/data/raw/movie_details/snapshot_date=2026-04-05/movie_id=467914.json"
```

```python
df = spark.read.option("multiline", "true").json(input_path)

# Optional: inspect schema
df.printSchema()
```

```python
# 3. Add snapshot_date
# ---------------------------
# For now, hard-code it. Later you can extract it from folder name like snapshot_date=2026-04-05
snapshot_date = "2026-04-05"

df = df.withColumn("snapshot_date", F.lit(snapshot_date).cast("date"))
```

```python
# 4. Core movie table
# ---------------------------
movies = (
    df.select(
        F.col("id").alias("movie_id"),
        "title",
        "original_title",
        "original_language",
        "overview",
        "homepage",
        "imdb_id",
        "release_date",
        "runtime",
        "status",
        "tagline",
        "adult",
        "video",
        "budget",
        "revenue",
        "popularity",
        "vote_average",
        "vote_count",
        "poster_path",
        "backdrop_path",
        "snapshot_date"
    )
)
movies.show()
```
